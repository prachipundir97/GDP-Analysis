## PartI-A: Plotting average % growth of states duration 2013-2016

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Reading from source: Local Path defined (Change path if needed to run somewhere else)
df_1a = pd.read_csv("C:/Users/Prachi Pundir/Desktop/GDP/state-wise-gdp.csv")

# Set index to Duration
# Remove the rows for Duration 2016-17 (GSDP current prices and % growth over previous year)
df_1a = df_1a.set_index('Duration')
df_1a = df_1a.drop('2016-17', axis=0)

#Taking transpose to calculate mean
df_1a = df_1a.T

# Calculate mean of average % growth for the states for % growth over previous years
df_1a['Average %Growth'] = df_1a.iloc[1:,6:].mean(axis=1)
df_1a = df_1a.sort_values('Average %Growth', ascending = False)
df_1a = df_1a.iloc[:-2,:]

#Plotting average% GSDP against each state
#plt.figure(figsize=(15,10))
#plt.xticks(rotation=90)
#sns.barplot(y='Average %Growth',x= df_1a.index, data= df_1a)

df_1a.plot.bar(figsize=(14,8),title= 'Average % Growth of States (2013-16)', y='Average %Growth')
plt.xlabel('States')
plt.show()
df_1a
```

### Treatment of missing values(NaN):

In the above bar graph , we have not treated NaN values, below mentioned are the reasons for it.

    a. If we fill Nan with 0, then for state such as , the mean would be taken by dividing the sum of % growth over 
       previous year by 3 and not by 2, resulting in wrong computation of mean for that state
    b. For state such as West Bengal, we there is no data for any year(NaN), this would result average as 0, but actually 
       that is not the case.

### Consistently Fast Growing States:

We deduced that states such as Andhra Pradesh, Rajasthan & Jharkhand have consistently growing percentage , every subsequent year, its GDP is rising by few points.

Alltogether the states such as Mizoram, Tripura & Nagaland are have top percentage combining all the years.

### Curosity Excercise:

My State is Delhi: Average % growth in Delhi is 12.16
If we take average of all the average % growth ,the result is 12.41.
Conclusion : Average %growth of Delhi is less than average of average % growth of all states, but the difference is very small.

## PartI-B : Plotting GDP Per Capita for all the states in the year 2014-15

```python
import pandas as pd
import numpy as np
import glob 
import matplotlib.pyplot as plt
import seaborn as sns

# Local Path defined (Change path if needed to run somewhere else)
my_path = r'C:/Users/Prachi Pundir/Desktop/GDP/GDP Indian States'
my_files = glob.glob(my_path+"\\*csv")

#combining all the state files into one dataframe
f_list = []
for fn in my_files:
    df_1b= pd.read_csv(fn,index_col=None,header=0,encoding='unicode_escape')
    df_1b= df_1b[['Item','2014-15']]
    df_1b=df_1b.T
    df_1b.columns=df_1b.iloc[0]
    df_1b=df_1b[1:]
    df_1b['state']=fn.split('\\')[-1].split('-')[1]
    f_list.append(df_1b)
    
frame_1b = pd.concat(f_list,axis=0,ignore_index=True, sort=True)
frame_1b


#Filtering data to get GDP Per capita 
frame_1b_Main = frame_1b.set_index('state')
#Removing UTs
frame_1b_Main = frame_1b_Main.drop(['Delhi','Puducherry','Chandigarh'])
frame_1b_percapita= frame_1b_Main[['Per Capita GSDP (Rs.)']]
frame_1b_percapita = frame_1b_percapita.sort_values(by="Per Capita GSDP (Rs.)",axis=0,ascending=False)
frame_1b_percapita

#Plotting Per Capita GDP against each state
plt.figure(figsize=(14,8))
plt.xticks(rotation=90)
sns.barplot(y='Per Capita GSDP (Rs.)',x= frame_1b_percapita.index, data= frame_1b_percapita)
plt.title('GDP Per Capita of all states for the year 2014-15')
plt.ylabel('Per Capita GSDP (Rs.)')
plt.xlabel('States')
plt.show()
```

## PartI-B : Plotting % contribution of the primary, secondary and tertiary sectors as a % of the total GDP for all the states for the year  2014-15:

```python
#selecting 3 sectors and GSDP of all the states
df_1b=frame_1b_Main[['Primary', 'Secondary','Tertiary','Gross State Domestic Product']] 


#calculating contribution of each sector
df_1b.loc[:,'primarycontribution'] = df_1b['Primary']/df_1b['Gross State Domestic Product']
df_1b.loc[:,'Secondarycontribution'] =df_1b['Secondary']/df_1b['Gross State Domestic Product']
df_1b.loc[:,'Tertiarycontribution']=df_1b['Tertiary']/df_1b['Gross State Domestic Product']
df_1b

#Computing total contribution from all 3 sectors
df_1b['Total_Contribution']=df_1b['primarycontribution']+df_1b['Secondarycontribution']+df_1b['Tertiarycontribution']
df_1b.sort_values(by='Total_Contribution',inplace=True,ascending=False)
df_1b.reset_index(inplace=True)
df_1b

df_1b_pc = df_1b[['state','primarycontribution','Secondarycontribution','Tertiarycontribution']]
df_1b_percontribution=df_1b_pc.set_index('state')

#plotting %contribution of each sector in stack bar graph against each state
df_1b_percontribution.plot.bar(stacked=True,width=0.90,figsize=(15,5))
plt.title('%contribution of primary,secondary and tertiary sectors to total GDP for all the states for the year 2014-15')
plt.ylabel( "%Contribution")
plt.xlabel('States')
plt.show()
```

## PartI-B: Plotting for C1, C2, C3, C4 groups of states based on GDP per capita for the year 2014-15:

```python
#Filtering sub sectors
frame_C=frame_1b_Main[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services',
               'Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting',
               'Financial services','Real estate, ownership of dwelling & professional services','Public administration',
               'Other services','TOTAL GSVA at basic prices','Taxes on Products','Subsidies on products','Gross State Domestic Product',
               'Per Capita GSDP (Rs.)']]


frame_Cg =frame_C.sort_values(by="Per Capita GSDP (Rs.)",axis=0,ascending=False)
frame_Cg.reset_index(inplace=True)
frame_Cg

#finding the quantile values
frame_group=frame_Cg['Per Capita GSDP (Rs.)'].quantile([0,0.2,0.5,0.85,1])
frame_group

#filling per capita GDP of quantiles as rule to categorize into C1,C2,c3,c4
rule= [(frame_Cg['Per Capita GSDP (Rs.)']>153246),
      (frame_Cg['Per Capita GSDP (Rs.)']<=153246)&  (frame_Cg['Per Capita GSDP (Rs.)']>104977),
      (frame_Cg['Per Capita GSDP (Rs.)']<=104977) & (frame_Cg['Per Capita GSDP (Rs.)']>65187),
      (frame_Cg['Per Capita GSDP (Rs.)']<=65187)]
category =['C1','C2','C3','C4']
frame_Cg['Category Group']=np.select(rule,category,default='')
frame_Cg

frame_groups=frame_Cg[['Agriculture, forestry and fishing','Mining and quarrying','Manufacturing','Electricity, gas, water supply & other utility services',
               'Construction','Trade, repair, hotels and restaurants','Transport, storage, communication & services related to broadcasting',
               'Financial services','Real estate, ownership of dwelling & professional services','Public administration',
               'Other services','Gross State Domestic Product','Category Group']]

frame_groups=frame_groups.set_index('Category Group')
frame_groups
```

```python
#C1 contribution to GDP
frame_category =frame_groups.loc['C1']
frame_C1= frame_category.sum(axis = 0, skipna = True)

df_C1 =pd.DataFrame(frame_C1)
df_C1.columns=['C1_total']
df_C1.loc[:,'C1(%)']=df_C1['C1_total']/101930677.0
df_C1g =df_C1.drop('Gross State Domestic Product')
df_C1g1 =df_C1g.sort_values(by=['C1(%)'],axis=0,ascending=False)

#cumsum() for 80% contribution
df_C1g2 =df_C1g1.cumsum()
df_C1_final = df_C1g2[0:6]
df_C1_final
df_C1_plot =df_C1g1[0:6]
df_C1_plot

#Plotting the percentage contribution against sub-sectors
sns.barplot(y= 'C1(%)',x=df_C1_plot.index, data = df_C1_plot )
plt.title('Sub-sectors contribution of 80% for C1 category')
plt.xticks(rotation=40)
plt.xlabel('Sub-sectors')
plt.ylabel('Contribution in %')
plt.show()
```
